```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
import geopandas as gpd
import folium
csv_file_path ="C:/Users/Parijat/Downloads/received_csv.csv"
```

```python
df = pd.read_csv(csv_file_path,names=['Name', 'latitude', 'longitude','No of Floors','Area(m^2)','Cost','Urban/Rural','lev','exc','struct','fit','paint','lev_time','exc_time','struct_time','fit_time','paint_time','total_time','%completion','Rate'])
df
df.columns = ['Name', 'latitude', 'longitude','No of Floors','Area(m^2)','Cost','Urban/Rural','lev','exc','struct','fit','paint','lev_time','exc_time','struct_time','fit_time','paint_time','total_time','%completion','Rate']
```

```python
def get_marker_color(rate1):
    if rate1 > 1:
        return 'green'
    elif rate1 == 1:
        return 'blue'
    else:
        return 'red'
```

```python
df['Marker Color'] = df['Rate'].apply(get_marker_color)
```

```python
map_center = [df['latitude'].mean(), df['longitude'].mean()]
mymap = folium.Map(location=map_center, zoom_start=12)

for index, row in df.iterrows():
    popup_text = f"Latitude: {row['latitude']}, Longitude: {row['longitude']}<br>Predicted Time: {row['total_time']} months,Rate: {row['Rate']}"
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=popup_text,
        icon=folium.Icon(color=row['Marker Color'])
    ).add_to(mymap)

# Display the map in the notebook
mymap
```
